```python
def get_dense(learning_rate=1e-3, num_output=2):
    l1, l2 = 0.1e-5, 0.5e-4
    inputs = keras.layers.Input(shape=(len(42),))
    hs = [512, 512, 512, 512]

    rates = [0.5, 0.5, 0.5, 0.15]

    x = inputs

    # dense layers
    for i, h in enumerate(hs):
        x = keras.layers.Dense(
            units=h,
            # kernel_regularizer=keras.regularizers.l2(l2)
        )(x)
        x = keras.layers.Activation('relu')(x)
        # x = keras.layers.BatchNormalization()(x)
        if i + 1 != len(hs):
            x = keras.layers.Dropout(rates[i])(x)
        else:
            x = keras.layers.Dropout(rates[i], name='last_dropout')(x)

    x = keras.layers.Dense(units=num_output, activation='softmax')(x)

    model = keras.models.Model(inputs=inputs, outputs=x)
    model.compile(
        optimizer=keras.optimizers.Adam(lr=learning_rate),
        loss=keras.losses.sparse_categorical_crossentropy,
        metrics=['accuracy']
    )
    return model
```

```python
def sort_standard(df, binary=True):
    df = df.sort_values(by='Full TOI ID', ascending=True)
    df = df.sort_values(by=['int_label', 'TIC'],
                   ascending=[False, True],
                   inplace=False, kind='mergesort')

    df = df.reset_index(drop=True)
    return df

def sort_df(df):
    df = df.sort_values(by=['int_label', 'ticid', 'tce_plnt_num'], 
                       ascending=[False, True, True],
                       inplace=False, kind='mergesort')
    df = df.reset_index(drop=True)
    return df

def sort_df_nolabel(df):
    df = df.sort_values(by=['ticid', 'tce_plnt_num'], 
                       ascending=[True, True],
                       inplace=False, kind='mergesort')
    df = df.reset_index(drop=True)
    return df
```
